```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X = np.random.normal(loc=65,scale=5,size=(10,))
```

```python
X
```

```python
m = 2

c = 1

Y_line = m*X + c
```

```python
Y_line
```

```python
plt.figure(figsize=(10,10))

line_obj = plt.plot(X,Y_line)

plt.setp(line_obj,linewidth=1.0,color='r',mec='g',dash_capstyle='round',marker='4',linestyle=':')

plt.xlabel("Random Data sampled from Normally Population with mean of 65 and std of 5")

plt.ylabel("Y = mX + c")
```

```python
Y_parabola = X**2
```

```python
plt.plot(X,Y_parabola,'b+')
```

```python
Y_cubic = X**3
```

```python
plt.plot(X,Y_cubic,'ro')
```

```python
plt.plot(X,Y_line,'bo',X,Y_parabola,'r+',X,Y_cubic,'g--')
```

```python
plt.figure(figsize=(15,15))

plt.subplot(221)

plt.plot(X,Y_line,'bo')

plt.xlabel("The x axis")

plt.ylabel("The Y axis")


plt.subplot(222)

plt.plot(X,Y_parabola,'r+')


plt.subplot(223)

plt.plot(X,Y_cubic,'g--')


plt.subplot(224)

plt.plot(X,(6*X+7),'r^')

plt.show()
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df_sliced = df.iloc[:,1:6]
```

```python
df_sliced['integer values'] = np.random.randint(low=0,high=6,size=(569,))
```

```python
df_sliced.drop(["area_mean"],inplace=True,axis=1)
```

```python
df_sliced
```

```python
plt.scatter('radius_mean','texture_mean',c='diagnosis',s='perimeter_mean',data=df_sliced)

plt.xlabel("radius mean")

plt.ylabel("texture mean")
```

```python
plot1,plot2 = plt.plot(df['radius_mean'],df['perimeter_mean'],'r+',df['area_mean'],df['texture_mean'],'g*',)

plt.setp(plot1,ms=2.0)

plt.setp(plot2,ms=5.0)

plt.title("Two Graphs, one between radisu Mean and Perimeter mean and another between area mean and texture mean")
```

```python
plt.plot(df['radius_mean'],df['texture_mean'])

plt.xlabel("Radius Mean")

plt.ylabel("Texture Mean")
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221)

plot1 = plt.scatter('radius_mean','texture_mean',c='diagnosis',s='texture_worst',data=df)

plt.xlabel("Radius mean")

plt.ylabel("texture Mean")

plt.title("Radius Mean v/s Texture Mean")



plt.subplot(224)

plt.scatter('perimeter_mean','area_mean',c='diagnosis',s='area_worst',data=df)

plt.xlabel("perimeter mean")

plt.ylabel("area mean")

plt.title("Perimeter Mean v/s Area Mean",)

plt.suptitle("plotting of two graphs in a diagonal manner")

plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.scatter('radius_mean','perimeter_mean',c='diagnosis',data=df)

plt.show()


plt.scatter('area_mean','texture_mean',c='diagnosis',data=df)

plt.show()
```

```python
fig,axes = plt.subplots(3,5,figsize=(25,25))

for i in range(0,3):
    
    for j in range(0,5):
            
        axes[i,j].scatter("radius_mean","texture_mean",c='diagnosis',s='perimeter_mean',data=df)
            
        axes[i,j].set_title("radius mean v/s texture mean",y=-0.2)
            
plt.suptitle("Graphs plotted at Diagonals",y=0.9)
```
